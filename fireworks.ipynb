FireWorks and atomate should be installed with the manual from https://materialsproject.github.io/fireworks/ and https://atomate.org/

Cif file should be ordered without doping sites.

wf_bandstructure add four workflows for one cif file

```python
import os
import numpy as np
from pymatgen.core import Structure
from fireworks import LaunchPad
from atomate.vasp.workflows.presets.core import wf_bandstructure
fn='./cif/'
for name in os.listdir(fn):
    struct = Structure.from_file(fn+name)  # note: many file formats supported, see function docs
    wf = wf_bandstructure(struct)
    lpad = LaunchPad.auto_load() # loads this based on the FireWorks configuration
    lpad.add_wf(wf)
```
